```python
def asearch(ret, freq, ext = '.parquet', add_risk_free = True):
    if ret in portfolios:
        df = globals()[ret]
        if  freq == 'D':
            if add_risk_free:
                df = pd.concat([df, nefin['Risk_free']], axis = 1)
            return df
        else: 
            df = resample_returns(df, freq)
            if add_risk_free:
                n = resample_returns(nefin, freq)['Risk_free']
                df = pd.concat([df, n], axis = 1)
            return df
    ret_dict = {
        
        'f': ['fac', 'factors'],
        'e': ['exp', 'expected']
    }
    path = f'{os.getcwd()}//Regression//{freq}//{ret_dict[ret][0]}//{ret_dict[ret][1]}{ext}'
    try: df = pd.read_parquet(path)
    except: df = pd.read_csv(path)
    df.index = pd.to_datetime(df.index)
    if add_risk_free:
        n = pd.DataFrame( columns = pd.MultiIndex.from_product([['Risk_free'], list(betas.keys())]))
        if  freq == 'D':
            for col in n.columns:
                n[col] = nefin['Risk_free'].shift()
        else:                
            for col in n.columns:
                n[col] = resample_returns(nefin, freq)['Risk_free'].shift()
    df = pd.concat(
        [df, n], axis = 1
    )
    return df

asearch('e', 'M')
```

```python
def eff_weights_dfs(portfolio:str, freq:str, cov:str, old = False) -> pd.DataFrame:
    if old:
        folder = f'{os.getcwd()}//Portfolios//Portfolios_old//'
    else: folder = f'{os.getcwd()}//Portfolios//'
    
    filename = f'{folder}eff_{portfolio}_{freq}_{cov}.parquet'
    return pd.read_parquet(filename)

def eff_performance_dfs(portfolio:str, freq:str, cov:str, old = False) -> pd.DataFrame:
    import warnings
    warnings.filterwarnings('ignore')
    try: df = eff_weights_dfs(portfolio, freq, cov, old)
    except: print(portfolio, freq, cov)
    cols = df.droplevel('stocks', axis = 1).columns.drop_duplicates()
    index = df.index
    df_returns = pd.DataFrame(np.nan, index = index, columns = cols)
    
    for col in cols:
        df_returns[col] = df[col].multiply(Search.search(portfolio, freq)).sum(axis = 1)
    return df_returns

def eff_concat_freq(portfolios: list , freq: str, covs: list, old = False) -> pd.DataFrame:
    list_of_dfs = [eff_performance_dfs(
        portfolio, freq, cov, old
    ) for portfolio in portfolios for cov in covs]
    df = pd.concat(list_of_dfs, axis = 1)
    if old:
        folder = f'{os.getcwd()}//Portfolios//Portfolios_old//'
    else: folder = f'{os.getcwd()}//Portfolios//'
    df.to_parquet(f'{folder}portfolios_raw_returns_{freq}.parquet')
    ((df+1).cumprod()).to_parquet(f'{folder}portfolios_raw_wealth_{freq}.parquet')

def join_column_names(df):
    ds = df.copy()
    if len(ds.columns.levels) > 1:
        ds.columns = ['_'.join(i) for i in ds.columns]
    return ds

def split_column_names(df, level_names = False):
    ds = df.copy()
    if type(df.columns[0]) == str:
        ds.columns = pd.MultiIndex.from_tuples([i.split('_') for i in ds.columns])
    if level_names:
        ds.columns.names = level_names
    return ds
```
