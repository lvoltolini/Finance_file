```python
df = adj 
years = [str(i) for i in range(2016, 2023)]

def table_year_sharpe(df, years: list) -> pd.DataFrame:
    sharpe = pd.DataFrame(columns = pd.MultiIndex.from_product([['Sharpe', 'Return', 'Risk', 'OBS'], years]))
    for year in years:
        bus_days = len(df.loc[year].index)**(1/2)
        ret = (df.loc[year]+1).prod()-1
        risk = df.loc[year].std()
        sharpe.loc[:, ('Sharpe', year)] = ret.div(risk*bus_days).replace(np.nan, 0)
        sharpe.loc[:, ('Return', year)] = ret
        sharpe.loc[:, ('Risk', year)] = risk
        sharpe.loc[:, ('OBS', year)] = bus_days
    return sharpe
# for_master = 3/19
def table_annualized_sharpe(df, num_of_years):
    annualized = pd.DataFrame()
    annualized['Return'] = (((df['Return']+1).prod(axis = 1))**(num_of_years)-1)
    annualized['Risk'] = (df['Risk']).mean(axis = 1)
    annualized['OBS'] = (df['OBS']**2).mean(axis = 1)**(1/2)
    annualized['Sharpe'] = annualized['Return']/(annualized['Risk']*annualized['OBS'])
    return annualized

def groupby_mapping(level_value):
    if level_value == 'br':
        return 'not_esg'
    else:
        return 'esg'
```
